```python
import numpy as np

# Read numeric grid from CSV (non-numeric entries become NaN)
raw = np.genfromtxt(filepath, delimiter=",", skip_header=1, dtype=float)

# First column = x coordinates, remaining columns = 2D height/error map
x_m = raw[:, 0]*1e-3  # convert from mm to m
H = raw[:, 1:]*1e-3  # convert from microns to meters

# First row (excluding first cell) = y coordinates
header = np.genfromtxt(filepath, delimiter=",", max_rows=1, dtype=str)
y_m = header[1:].astype(float)*1e-3  # convert from mm to m
```

```python
# x_mm is reverted because it is supposed to be increasing
Shadow.ShadowTools.write_shadow_surface(H.T, -x_m, y_m, outFile = '../data/ssrl_bl10-1_M0_mirror_Ux_Data_02-25-26.dat') 
```
